```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
iris = pd.read_csv("https://gist.githubusercontent.com/netj/8836201/raw/6f9306ad21398ea43cba4f7d537619d0e07d5ae3/iris.csv")
iris

# Création des trois DataFrames CAR MANQUANT DANS LE fichier.csv

# FIGURE 4/4 M03 - Les librairies Python pour le ML.pdf
iris_setosa = iris[iris["variety"] == "Setosa"]

iris_versi = iris[iris["variety"] == "Versicolor"]

iris_virginica = iris[iris["variety"] == "Virginica"]
fig, axes = plt.subplots(2, 3, figsize=(10, 10))

# ------------------------
# Ligne 1 : Sépales
# ------------------------

sub1 = axes[0, 0]
sub1.plot(
    iris_setosa['sepal.length'],
    iris_setosa['sepal.width'],
    linestyle='',
    marker='o'
)
sub1.set_ylabel('Sepal width')
sub1.set_title('Setosa')

sub2 = axes[0, 1]
sub2.plot(
    iris_virginica['sepal.length'],
    iris_virginica['sepal.width'],
    linestyle='',
    marker='o'
)
sub2.set_title('Virginica')
sub2.set_xlabel('Sepal length')

sub3 = axes[0, 2]
sub3.plot(
    iris_versi['sepal.length'],
    iris_versi['sepal.width'],
    linestyle='',
    marker='o'
)
sub3.set_title('Versicolor')

# ------------------------
# Ligne 2 : Pétales
# ------------------------

sub4 = axes[1, 0]
sub4.plot(
    iris_setosa['petal.length'],
    iris_setosa['petal.width'],
    linestyle='',
    marker='o'
)
sub4.set_ylabel('Petal width')

sub5 = axes[1, 1]
sub5.plot(
    iris_virginica['petal.length'],
    iris_virginica['petal.width'],
    linestyle='',
    marker='o'
)
sub5.set_xlabel('Petal length')

sub6 = axes[1, 2]
sub6.plot(
    iris_versi['petal.length'],
    iris_versi['petal.width'],
    linestyle='',
    marker='o'
)

plt.show()
```
